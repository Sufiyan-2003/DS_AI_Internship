TASK-1: The Paradigm Shift (Traditional vs. Machine Learning)

Machine Learning is a branch of artificial intelligence where computers learn patterns from data instead of being explicitly programmed with fixed rules. 
In traditional programming, we give the computer both the rules and the data to get results. But in Machine Learning, we provide data along with the correct answers, and the system automatically discovers the rules by identifying patterns. 
This allows machines to improve their performance over time and make predictions or decisions on new, unseen data. For example, instead of manually writing rules to detect spam emails, a machine learning model learns from thousands of labeled emails and figures out what makes an email spam.

TASK-2: The Three Branches of ML

1. Supervised Learning

Supervised Learning is a type of machine learning where the data is labeled, meaning each input has a correct output. The model learns from these input-output pairs and predicts outcomes for new data.

Real-World Examples:

Spam Email Detection, 
House Price Prediction, 
Student Result Prediction, 
Disease Diagnosis.

2. Unsupervised Learning

Unsupervised Learning works with data that has no labels. The model tries to find hidden patterns, groupings, or structures within the data.

Real-World Examples:

Customer Segmentation in marketing, 
Grouping similar news articles, 
Market Basket Analysis, 
Anomaly Detection in transactions.

3. Reinforcement Learning

Reinforcement Learning is where an agent learns by interacting with an environment and receiving rewards or penalties. The goal is to maximize long-term reward.

Real-World Examples:

Self-driving cars, 
Game-playing AI, 
Robot navigation, 
Recommendation systems that improve over time.
